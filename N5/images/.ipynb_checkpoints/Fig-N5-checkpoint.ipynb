```python
#Ex-1.12
def grafico_f_tan(f, p, intervalox, intervaloy, ticks):
    
    pf = plot(f, intervalox, aspect_ratio=1)
    df = diff(f, x)
    tangente = df(x=p)*(x-p) + f(x=p)
    pt = plot(tangente, intervalox, aspect_ratio=1, color='red', ymin=intervaloy[0], ymax=intervaloy[1], 
              figsize=[4,4], thickness=1.5,axes_labels=['$x$','$y$'], 
              ticks= ticks)
    r = (intervalox[1] - intervalox[0])/150
    c = circle((p, f(x=p)), r, color='green', thickness=1.5)
    
    return (pf+pt+c)
p = grafico_f_tan(x^2, 1, (-1, 2), (-1/2, 2), ticks=[[1],[1]]) 
p+= text(r"$y=x^2$", (1/2, (1.2)^2 ), fontsize=12)
p+= text(r"$y=2x-1$", (1.4, 1/2 ), color='red',fontsize=12)
#p
```

```python
# Ex-1.18
def grafico_f_tan(f, p, intervalox, intervaloy, ticks):
    
    pf = plot(f, intervalox, aspect_ratio=1, exclude=[0])
    df = diff(f, x)
    tangente = df(x=p)*(x-p) + f(x=p)
    pt = plot(tangente, intervalox, aspect_ratio=1, color='red', ymin=intervaloy[0], ymax=intervaloy[1], 
              figsize=[4,4], thickness=1.5,axes_labels=['$x$','$y$'], 
              ticks= ticks)
    r = (intervalox[1] - intervalox[0])/150
    c = circle((p, f(x=p)), r, color='green', thickness=1.5)
    
    return (pf+pt+c)
f = 3*x^2-2*x
p = grafico_f_tan(f, 1, (-1/2, 1.25), (-1/2, 2), ticks=[[1],[1]]) 
p+= text(r"$y={}$".format(latex(f)), (1/2-0.1, 1/2), fontsize=12)
p+= text(r"$y=4x-3$", (1.2, 1/5 ), color='red',fontsize=12)
p
```
